```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from math import log
```

```
def logloss(true_label, predicted):
    eps = 1e-10
    max = 1.0 - eps

    p = np.clip(predicted, eps, max)

    if true_label == 1:
        return -log(np.clip(p, eps, max))
    else:
        return -log(np.clip(1-p, eps, max))
```

```
def plot_logloss(t1, y):
    plt.figure(1, figsize=(15,10))
    plt.plot(t1,list(map(logloss, y, t1)), color="black")
    plt.title("Log-Loss vs predict probability")
    plt.xlabel("predict probability")
    plt.ylabel("log-loss")

    p005 = 0.005
    plt.scatter([p005], logloss(1, p005), s=50, marker='o', color='red')
    plt.annotate(r'0.005',
                  xy=(p005, logloss(1, p005)), xycoords='data',
                  xytext=(+20, +20), textcoords='offset points', fontsize=15,
                  arrowprops=dict(facecolor='red', shrink=0.05, connectionstyle="arc3, rad=.2"), color='red')
    
    p001 = 0.00001
    plt.scatter([p001], logloss(1, p001), s=50, marker='o', color='red')
    plt.annotate(r'0.00001',
                  xy=(p001, logloss(1, p001)), xycoords='data',
                  xytext=(+20, +20), textcoords='offset points', fontsize=15,
                  arrowprops=dict(facecolor='red', shrink=0.05, connectionstyle="arc3, rad=.2"), color='red')
    
    
    plt.show()
    
t = np.arange(0.000001, 1, 0.001)
yt = [1] * len(t)

plot_logloss(t, yt)
```

> trick for logloss: 当概率极小时会导致loss变大
