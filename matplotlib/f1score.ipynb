```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib
```

```
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)

X = np.arange(0.0001, 1, 0.01)
Y = np.arange(0.0001, 1, 0.01)
X, Y = np.meshgrid(X, Y)
mean = (X+Y)/2
f1_score = 2*(X*Y)/(X+Y)

ax.set_xlabel('recall')
ax.set_ylabel('precision')
ax.set_zlabel('evaluation metric')


ax.plot_surface(X, Y, mean, alpha = 0.8, cmap='Greys')
ax.plot_surface(X, Y, f1_score, rstride=1, cstride=1,alpha = 0.5, cmap='rainbow')
ax.text(0, 0.9, 0.2, "F1-score")
ax.text(0, 0.9, 0.6, "Mean")
ax.view_init(30,120)
ax.invert_xaxis()
plt.show()
```

```
plt.plot(mean)
plt.title('mean')
```

```
plt.plot(f1_score)
plt.title('f1-score')
```
