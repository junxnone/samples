```
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```
plt.figure(figsize=(16,8))

m = Basemap()
m.drawcoastlines()

plt.show()
```

```
plt.figure(figsize=(16,8))

m = Basemap()
m.drawcoastlines()
m.drawcountries(linewidth=1.5)

plt.show()
```
