```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.arange(-10,10,0.1)
s=np.exp(t)
plt.subplot(2,2,1)
plt.plot(t,s,'b--')
plt.ylabel('y1')
plt.subplot(2,2,2)
plt.plot(2*t,s,'r--')
plt.ylabel('y2')
plt.subplot(2,2,3)
plt.plot(3*t,s,'m--')
plt.subplot(2,2,4)
plt.plot(4*t,s,'k--')
plt.show()
```

```python
plt.figure(figsize=(10,10))
for i in range(4):
  plt.subplot(2,2,i+1)
  s=np.exp(t/(i+1))
  plt.plot(t,s,'b--')
  plt.ylabel('y1')
  plt.title(f'T={i+1}')
```
