```python
import cv2
import numpy as np
import imageio
import imgaug as ia
import matplotlib.pyplot as plt
%matplotlib inline

def plot_hv_std(input_path):
    img = cv2.imread(input_path)
    plt.figure(figsize=(15,15))
    plt.imshow(img)
    test_data = np.asarray(img)
    h_std = np.std(test_data, axis = (0,2))
    v_std = np.std(test_data, axis = (1,2))
    print(h_std.shape)
    print(v_std.shape)

    plt.figure(figsize=(10,5))
    h_plt, = plt.plot(np.arange(1, len(h_std)+1), h_std, 'b')

    plt.title('horizontal standard deviation') 
    plt.xlabel('pixel')    
    plt.ylabel('std')      

    plt.figure(figsize=(10,5))
    v_plt, = plt.plot(np.arange(1, len(v_std)+1), v_std, 'r')
    plt.title('vertical standard deviation') 
    plt.xlabel('pixel')     
    plt.ylabel('std')      

input_path = "../opencv/img/t.png"
plot_hv_std(input_path)
```
