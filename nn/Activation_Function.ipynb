## Logistic (sigmoid)


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(-5, 5, 0.01)
y = 1 / (1 + np.exp(-x))
dy = y * (1-y)
plt.plot(x,y,color='b',label='sigmoid')
plt.plot(x,dy,color='g', label='sigmoid\'')
plt.title('Logistic/sigmoid Activation Function')
plt.legend()
plt.grid()
plt.xlabel('Input')
plt.ylabel('Output');
```

## Hyperbolic Tangent - tanh

```python
x = np.arange(-5, 5, 0.01)
y = (2 / (1 + np.exp(-2*x)))-1
dy = 1 - (y) ** 2
plt.plot(x,y,color='b',label='tanh')
plt.plot(x,dy,color='g',label='tanh\'')
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output');
plt.grid()
plt.legend()
```

## ReLU

```python
x = np.arange(-5, 5, 0.01)
z = np.zeros(len(x))
y = np.maximum(z,x)
dy = y/y+1e-9
plt.plot(x,y,color='b',label='ReLU')
plt.plot(x,dy,color='g',label='ReLU\'')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output');
plt.grid()
plt.legend()
```

## Softplus

```python
x = np.arange(-5, 5, 0.01)
y = np.log(1+np.exp(x))
plt.plot(x,y,color='b',label='Softplus')
plt.title('Softplus Activation Function')
plt.xlabel('Input')
plt.ylabel('Output');
plt.grid()
plt.legend()
```
