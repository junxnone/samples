```
import torch  
from torchvision.datasets import *
import matplotlib.pyplot as plt
```

```
def plot_ds(dataset, N=8):
  i = 1
  plt.figure(figsize=(20, 5 * N//4 + 1))
  for data in dataset:
      inputs, labels = data
      plt.subplot(N//4 + 1,4,i)
      plt.imshow(inputs)
      i += 1
      if i > N:
          break
```

```
CelebA_DS = CelebA('./', download=True)
plot_ds(CelebA_DS)
```

```
CIFAR10_DS = CIFAR10('./', download=True)
plot_ds(CIFAR10_DS)
```

```
CIFAR100_DS = CIFAR100('./', download=True)
plot_ds(CIFAR100_DS)
```
