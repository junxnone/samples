```python
import torch
import matplotlib.pyplot as plt
```

```python
def plot_scheduler(lr_scheduler_fn, lr_scheduler_name):
  lr_list = []
  for i in range(100):
    lr_list +=lr_scheduler_fn.get_lr()
    lr_scheduler_fn.step()
  plt.figure(figsize=(12,7))
  plt.plot(list(range(len(lr_list))), lr_list)
  plt.legend([lr_scheduler_name], fontsize=20)
  plt.xlabel('epoch', size=15)
  plt.ylabel('lr', size=15)
  plt.show()
```

# CosineAnnealingLR

```python
lr = 0.001
scheduler_cos = torch.optim.lr_scheduler.CosineAnnealingLR(torch.optim.SGD([torch.ones(1)], lr), 20, 0)
plot_scheduler(scheduler_cos, 'CosineAnnealingLR')
```

# StepLR

```python
scheduler_step = torch.optim.lr_scheduler.StepLR(torch.optim.SGD([torch.ones(1)], lr), 20, 0.5)
plot_scheduler(scheduler_step, 'StepLR')
```

# ExponentialLR

```python
scheduler_exp = torch.optim.lr_scheduler.ExponentialLR(torch.optim.SGD([torch.ones(1)], lr), 0.95)
plot_scheduler(scheduler_exp, 'ExponentialLR')
```

# CyclicLR

```python
scheduler_cyc = torch.optim.lr_scheduler.CyclicLR(torch.optim.SGD([torch.ones(1)], lr), 0.0001, 0.001, 20)
plot_scheduler(scheduler_cyc, 'CyclicLR')
```

# MultiStepLR

```python
scheduler_mul = torch.optim.lr_scheduler.MultiStepLR(torch.optim.SGD([torch.ones(1)], lr), [20, 30, 60, 80], 0.8)
plot_scheduler(scheduler_mul, 'MultiStepLR')
```

# MultiplicativeLR

```python
lmbda = lambda epoch: 0.95
scheduler_mulpa = torch.optim.lr_scheduler.MultiplicativeLR(torch.optim.SGD([torch.ones(1)], lr), lr_lambda=lmbda)
plot_scheduler(scheduler_mulpa, 'MultiplicativeLR')
```

# OneCycleLR

```python
scheduler_oc = torch.optim.lr_scheduler.OneCycleLR(torch.optim.SGD([torch.ones(1)], 0.00005), max_lr=0.001, steps_per_epoch=20, epochs=100)
def plot_scheduler_batch(lr_scheduler_fn, lr_scheduler_name):
  lr_list = []
  for i in range(100):
    for bi in range(20):
      lr_list +=lr_scheduler_fn.get_lr()
      lr_scheduler_fn.step()
  plt.figure(figsize=(12,7))
  plt.plot(list(range(len(lr_list))), lr_list)
  plt.legend([lr_scheduler_name], fontsize=20)
  plt.xlabel('epoch', size=15)
  plt.ylabel('lr', size=15)
  plt.show()
  
  
plot_scheduler_batch(scheduler_oc, 'OneCycleLR')
```

# CosineAnnealingWarmRestarts

```python
scheduler_cawr = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(torch.optim.SGD([torch.ones(1)], lr), 20, 2)
plot_scheduler(scheduler_cawr, 'CosineAnnealingWarmRestarts')
```
