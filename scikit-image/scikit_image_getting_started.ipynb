```
import skimage
from skimage import data, io, filters
import matplotlib.pyplot as plt
import matplotlib
```

```
print(skimage.__version__)
```

```
image = data.coins()
# ... or any other NumPy array!
edges = filters.sobel(image)
io.imshow(image)
```

```
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

images = data.stereo_motorcycle()
ax[0].imshow(images[0])
ax[1].imshow(images[1])

fig.tight_layout()
plt.show()
```

```
fig, axes = plt.subplots(4, 5, figsize=(10, 10))
ax = axes.ravel()
images = data.lfw_subset()
for i in range(20):
    ax[i].imshow(images[90+i], cmap=plt.cm.gray)
    ax[i].axis('off')
fig.tight_layout()
plt.show()
```

```
images = ('astronaut',
          'binary_blobs',
          'camera',
          'checkerboard',
          'chelsea',
          'clock',
          'coffee',
          'coins',
          'horse',
          'logo',
          'page',
          'text',
          'rocket',
          'hubble_deep_field',
          'immunohistochemistry',
          'microaneurysms',
          'moon',
          'retina',
          )
fig, axes = plt.subplots(3, 6, figsize=(15, 10))
ax = axes.ravel()
i = 0
for name in images:
    caller = getattr(data, name)
    image = caller()
    if image.ndim == 2:
        ax[i].imshow(image, cmap=plt.cm.gray)
    else:
        ax[i].imshow(image)
    ax[i].set_title(name)
    ax[i].axis('off')
    i += 1
fig.tight_layout()
plt.show()
```
