```python
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
digits = datasets.load_digits()
print(digits.data.shape)
print(digits.target.shape)
fig = plt.figure(figsize=(10,10))

for i in range(100):
    ps = fig.add_subplot(10,10,i+1, xticks=[], yticks=[])
    ps.imshow(digits.images[i], cmap='gray')
```

```python
iris = datasets.load_iris()
classes_list = ['Setosa', 'Versicolour', 'Virginica']
print(iris.data.shape)
print(iris.target.shape)
print('number | 花萼长度 | 花萼宽度 | 花瓣长度 | 花瓣宽度 | classes')
print('-- | -- | -- | -- | -- | --')
for i in range(0,iris.data.shape[0]):
    print(str(i) + ' | ', end = ' ')
    for t in iris.data[i]:
        print(str(t) + ' | ', end = ' ')
    print(classes_list[iris.target[i]])
```
