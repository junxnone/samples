## Logistic (sigmoid)


```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt

# Generating data to plot
x_data = np.linspace(-6,6,100)
y_data = tf.nn.sigmoid(x_data)

# Plotting
plt.plot(x_data, y_data)
plt.title('Sigmoid Function')
plt.grid()
plt.show()
```

## Tanh

```python
x_data = np.linspace(-6,6,100)
y_data = tf.nn.tanh(x_data)

# Plotting
plt.plot(x_data, y_data)
plt.title('tanh')
plt.grid()
plt.show()
```

## ReLU

```python
x_data = np.linspace(-6,6,100)
y_data = tf.nn.relu(x_data)

# Plotting
plt.plot(x_data, y_data)
plt.title('RELU')
plt.grid()
plt.show()
```

## Leaky ReLU

```python
x_data = np.linspace(-6,6,100)
y_data = tf.nn.leaky_relu(x_data)

# Plotting
plt.plot(x_data, y_data)
plt.title('leaky_relu Function')
plt.grid()
plt.show()
```

## SELU

```python
x_data = np.linspace(-6,6,100)
y_data = tf.nn.selu(x_data)

# Plotting
plt.plot(x_data, y_data)
plt.title('selu Function')
plt.grid()
plt.show()
```

## ELU

```python
x_data = np.linspace(-6,6,100)
y_data = tf.nn.elu(x_data)

# Plotting
plt.plot(x_data, y_data)
plt.title('elu Function')
plt.grid()
plt.show()
```

## SoftSign

```python
x_data = np.linspace(-6,6,100)
y_data = tf.nn.softsign(x_data)

# Plotting
plt.plot(x_data, y_data)
plt.title('softsign Function')
plt.grid()
plt.show()
```
